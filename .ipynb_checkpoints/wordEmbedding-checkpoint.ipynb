```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
# Vocabulary size
voc_size=10000
```

```python
#One Hot Representation
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

```python
#Word Embedding Represntation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8 #making all same size of 8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) #padding=pre beacuse adding 0's at front
print(embedded_docs)
```

```python
dim=10 #no. of dimension or features to represent
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```

```python
print(model.predict(embedded_docs)[0]) #we get vector for every word in the sentence and size of vector is 10 as we defined
```
